# Activation Functions: Adding Non-Linearity to Neural Networks

## **Introduction**
Activation functions are a crucial component of neural networks. They introduce **non-linearity** into the model, allowing it to learn complex patterns and relationships in the data. Without activation functions, a neural network would simply be a linear regression model.

## **Key Activation Functions**
- **Sigmoid:** Maps input values to a range between 0 and 1. Commonly used in binary classification.
- **ReLU (Rectified Linear Unit):** Returns the input directly if it is positive; otherwise, it returns 0. Widely used in hidden layers.
- **Tanh (Hyperbolic Tangent):** Maps input values to a range between -1 and 1. Often used in hidden layers.

## **Why Are They Important?**
- Enable neural networks to model complex, non-linear relationships.
- Help in achieving better performance and faster convergence during training.

## **Implementation**
In this notebook, we will:
1. Implement the Sigmoid, ReLU, and Tanh activation functions.
2. Visualize their behavior using Matplotlib.

Let's dive in! ðŸ§ 

```python
import numpy as np
import matplotlib.pyplot as plt

# Define activation functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

def tanh(x):
    return np.tanh(x)

# Plot activation functions
x = np.linspace(-10, 10, 100)
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.plot(x, sigmoid(x))
plt.title("Sigmoid")
plt.grid()

plt.subplot(1, 3, 2)
plt.plot(x, relu(x))
plt.title("ReLU")
plt.grid()

plt.subplot(1, 3, 3)
plt.plot(x, tanh(x))
plt.title("Tanh")
plt.grid()

plt.show()
```
